# **Finding top 5 frequent species in 5 cities:**

---



```python
# Count the repetitions of unique values in the 'scientific_name' column
value_counts = combined_df['scientific_name'].value_counts()

# Print the unique values and their counts
for value, count in value_counts.items():
    print(f"{value}: {count}")
```

```python
# Show the top five unique values and their counts
top_five = value_counts.head(5)
print(top_five)
```

# **The number of repetition of 5 species in SanDiego city:**

```python
scientific_name = 'scientific_name'
top_five_values = ['Platanus acerifolia','Syagrus romanzoffianum','Washingtonia robusta','Magnolia grandiflora','Liquidambar styraciflua']

# Create a dictionary to store the counts
value_counts1 = {value: (df1[scientific_name] == value).sum() for value in top_five_values}

species_to_count = []
new_row = {'city':'SanDiego'}
# Print the counts
for value, count in value_counts1.items():
    species_to_count.append(value)
    new_row[value] = count
    new_row['total'] = total_count1
    print(f"{value}: {count}")

new_columns = ['city','scientific_name']
new_df1 = df1[new_columns]

result_df = pd.DataFrame(columns=['city'] + species_to_count + ['total'])
result_df = result_df.append(new_row, ignore_index=True)
result_df
```
